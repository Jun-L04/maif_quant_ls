```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
bm = []
n = 5000
B_s = 0
x = np.linspace(0, n, n)
for t in range(len(x)):
    B_t = np.random.normal(0,1)
    bm.append(B_s + B_t)
    B_s = B_s + B_t
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x,bm)
plt.title("Brownian Motion")
plt.show()
```
